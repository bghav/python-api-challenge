```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import json
import gmaps
import os
import gmplot 
import googlemaps
import pprint
import time

# Import API key
from api_keys import g_key
# Access maps with unique API key
gmaps.configure(api_key=g_key)

# File to Load (Remember to Change These)
wmap = "Resources/output_data.csv"

# Read Purchasing File and store into Pandas data frame
weather_data = pd.read_csv(wmap)
```

```python
weather_data
```

```python
# Heatmap of humidity
locations = weather_data[["Lat", "Lng"]]
humidity = weather_data["Humidity"]
#gmap = gmplot.GoogleMapPlotter(0, 0, 2)
fig = gmaps.figure()
heat_layer = gmaps.heatmap_layer(locations, weights=humidity, dissipating=False, max_intensity=300, point_radius=5)
fig.add_layer(heat_layer)

fig

```

```python
# geocoordinates
target_coordinates = "17.4924, 44.1277"
target_search = "Lodging"
target_radius = 50000
target_type = "Lodging"

# set up a parameters dictionary
params = {
    "location": target_coordinates,
    "keyword": target_search,
    "radius": target_radius,
    "type": target_type,
    "key": g_key
}

# base url
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

# run a request using our params dictionary
response = requests.get(base_url, params=params)

# convert response to json
places_data = response.json()

# Print the json (pretty printed)
print(json.dumps(places_data, indent=4, sort_keys=True))
```

```python
# Print the name and address of hotel that appears
print(places_data["results"][0]["name"])
print(places_data["results"][0]["vicinity"])
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel_Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Lat", "Lng"]]
```

```python
# Create bank symbol layer
hotel_layer = gmaps.symbol_layer(
    locations, fill_color='rgba(100, 150, 100, 0.4)',
    stroke_color='rgba(0, 0, 150, 0.4)', scale=2,
    info_box_content=[f"Hotel: {Hotel_Name}" for Hotel_Name in hotel_df.iterrows()]
)


fig = gmaps.figure()

fig.add_layer(heat_layer)
fig.add_layer(hotel_layer)

fig
```
